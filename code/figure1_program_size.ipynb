# Figure 1: Number of male and female MSTP students by academic year (AY)

This notebook uses Python3 and imports all necessary dependencies as it is run. 

**Input**: Takes in the raw datafile entitled "matriculants_raw_data.csv" in the `data` directory.  
**Output**: Saves a PDF of figure 3 to `figures` directory. 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# import raw data into a Pandas dataframe 
DF = pd.read_csv('../data/program_size_raw_data.csv') # this path may need to be updated depending on where you running this script from and where you have the raw datafile saved
DF.rename(columns={'AY':'academic year', 'F':'Female', 'M':'Male'}, inplace=True)
# DF = DF.set_index('academic year')
```

```python
# set plot basic parameters
axes = plt.gca()
%matplotlib inline  
plt.figure(figsize=(12,4))
plt.xticks(rotation=45)

mpl.rcParams['axes.spines.left'] = True
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.bottom'] = True

plt.plot('academic year', 'Female', data=DF, marker='o', color='#B3B3B3')
plt.plot('academic year', 'Male', data=DF, marker='s', color='#000000')
plt.ylabel('Number of students', fontsize=12)
plt.legend(fontsize=12, frameon=False, bbox_to_anchor=(0.2, 1))


# Save plot as a PDF
plt.savefig('../figures/figure1_matriculants-by-year.pdf', dpi=300, transparent=True, bbox_inches='tight')
```
